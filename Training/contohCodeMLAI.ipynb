# Materi 4

```python
import numpy as np
#mapel1-3
nilai = np.array([
    [78, 85, 90],
    [92, 88, 79],
    [75, 80, 85],
    [89, 91, 94]
])

print("Data Nilai Siswa:\n", nilai)
print("Shape array:", nilai.shape)
```

```python
#ambil nilai mapel1
mapel1 = nilai[:, 0]
print("Nilai Mapel 1:\n", mapel1)

#ambil nilai siswa2
siswa2 = nilai[1,:]
print("Nilai Siswa 2:\n", siswa2)

#ambil nilai mapel 3 dari 2 siswa
mapel3_siswa2 = nilai[2:,2]
print("Nilai Mapel 3 Siswa 2:\n", mapel3_siswa2)
```

```python
mapel2 = nilai[:,1]

mapel2_sort = np.sort(mapel2)

print("Nilai Mapel 2 Asli:\n", mapel2)
print("Nilai Mapel 2 Asc:\n",mapel2_sort)
print("Nilai Mapel 2 Desc:\n",mapel2_sort[::-1])
```

```python
print("PROJECT: Ranking Siswa berdasarkan Mata Pelajaran 2")

# ambil index urut
ranking = np.argsort(mapel2)[::-1]
ranking_nilai = mapel2[ranking]

for i, index in enumerate(ranking):
    print(f"Peringkat {i+1}: Siswa {index} dengan nilai Mata Pelajaran 2 = {ranking_nilai[i]}")

# rata-rata mapel2
print("\nRata-rata nilai Mata Pelajaran 2:", mapel2.mean())
```

# Materi 5

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42) #random nya biar sama

penjualan = np.linspace(10,50,12) * np.random.random(12) #noise

bulan = np.array(["Jan", "Feb", "Mar", "Apr", "Mei", "Jun",
                  "Jul", "Agu", "Sep", "Okt", "Nov", "Des"])

# Perhitungan statistik
total = np.sum(penjualan)
rata_rata = np.mean(penjualan)
deviasi = np.std(penjualan)

# Visualisasi
plt.figure(figsize=(10, 5)) #x y
label_stat = (
    f"Total: {total} juta\n"
    f"Rata-rata: {rata_rata:.2f} juta\n"
    f"Deviasi: {deviasi:.2f} juta"
)

line, = plt.plot(bulan, penjualan, marker='o')
plt.legend([line], [label_stat], loc='upper left', frameon=True)

plt.title("Penjualan Bulanan Tahun 2025")
plt.xlabel("Bulan")
plt.ylabel("Penjualan (juta rupiah)")
plt.grid(True)
plt.show()


```

# Materi 6

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

days = np.arange(1, 31)
units_sold = np.random.randint(5, 50, size=30)           # unit terjual
price_per_unit = np.random.randint(10000, 50000, size=30)  # harga per unit

# Gabungkan dalam 2D array
sales_data = np.column_stack((days, units_sold, price_per_unit))
```

```python
df = pd.DataFrame(
    sales_data,
    columns=["Hari", "Unit Terjual", "Harga Unit"]
)

# Buat kolom baru: Total Pendapatan
df["Total Pendapatan"] = df["Unit Terjual"] * df["Harga Unit"]

print("DataFrame Pandas:")
print(df.head(), "\n")

# Analisis Statistik
total_revenue = df["Total Pendapatan"].sum()
avg_units = df["Unit Terjual"].mean()
max_day = df.loc[df["Total Pendapatan"].idxmax()]

print("Total Pendapatan 30 Hari:", total_revenue)
print("Rata-rata Unit Terjual per Hari:", round(avg_units, 2))
print("\nHari dengan Pendapatan Tertinggi:")
print(max_day, "\n")
```

```python
# 1. Penjualan unit per hari (line plot)
plt.figure()
plt.plot(df["Hari"], df["Unit Terjual"], marker="o")
plt.title("Unit Terjual per Hari")
plt.xlabel("Hari")
plt.ylabel("Unit")
plt.grid(True)
plt.show()

# 2. Total pendapatan per hari (bar chart)
plt.figure()
plt.bar(df["Hari"], df["Total Pendapatan"])
plt.title("Total Pendapatan per Hari")
plt.xlabel("Hari")
plt.ylabel("Pendapatan (Rp)")
plt.tight_layout()
plt.show()

# 3. Distribusi harga unit (histogram)
plt.figure()
plt.hist(df["Harga Unit"], bins=10)
plt.title("Distribusi Harga Unit")
plt.xlabel("Harga Unit (Rp)")
plt.ylabel("Frekuensi")
plt.tight_layout()
plt.show()
```

# Materi 7

```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter
import pandas as pd
import matplotlib.pyplot as plt


# Set the path to the file you'd like to load
file_path = "police.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "melihkanbay/police",
  file_path,
  # Provide any additional arguments like
  # sql_query or pandas_kwargs. See the
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

df.info()
```

```python
# gabungkan kolom tanggal & waktu menjadi datetime
df['stop_datetime'] = pd.to_datetime(df['stop_date'] + " " + df['stop_time'])

# ekstraksi agregasi waktu
df['year']  = df['stop_datetime'].dt.year
df['month'] = df['stop_datetime'].dt.to_period('M')
df['hour']  = df['stop_datetime'].dt.hour


print("Jumlah data:", len(df))
print("\nDistribusi gender:\n", df['driver_gender'].value_counts(dropna=False))
print("\nDistribusi pelanggaran:\n", df['violation'].value_counts(dropna=False))
print("\nRasio search conducted:", df['search_conducted'].mean())
print("\nRasio arrested:", df['is_arrested'].mean())
```

```python
vc = df['violation'].value_counts()

plt.figure(figsize=(8,4))
plt.bar(vc.index, vc.values)
plt.title("Distribusi Jenis Pelanggaran")
plt.xlabel("Jenis Pelanggaran")
plt.ylabel("Jumlah")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("\n\n\n\n")

g = df['driver_gender'].value_counts()

plt.figure(figsize=(4,4))
plt.bar(g.index, g.values)
plt.title("Distribusi Gender Pengemudi")
plt.xlabel("Gender")
plt.ylabel("Jumlah")
plt.tight_layout()
plt.show()

print("\n\n\n\n")

arrest_rate = df.groupby('violation')['is_arrested'].mean()

plt.figure(figsize=(8,4))
plt.bar(arrest_rate.index, arrest_rate.values)
plt.title("Rasio Penangkapan per Jenis Pelanggaran")
plt.ylabel("Rasio")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("\n\n\n\n")

df['driver_age'] = pd.to_numeric(df['driver_age'], errors='coerce')
df_age = df.dropna(subset=['driver_age'])

plt.figure(figsize=(8,4))
plt.hist(df_age['driver_age'], bins=30)
plt.title("Distribusi Usia Pengemudi")
plt.xlabel("Usia")
plt.ylabel("Frekuensi")
plt.tight_layout()
plt.show()
```
