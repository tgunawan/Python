## 1ï¸âƒ£ Filtering & Seleksi Data

```python
import pandas as pd

data = {
    "Nama": ["Andi", "Budi", "Citra", "Doni"],
    "Umur": [20, 22, 19, 23],
    "Nilai": [80, 90, 70, 85]
}

df = pd.DataFrame(data)
```

```python
df_lulus = df[df["Nilai"] >= 80]
print(df_lulus)
```

## 2ï¸âƒ£ Aggregation (Agregasi Data)

## 5ï¸âƒ£ Mini Project: Analisis Nilai Siswa

### ğŸ¯ Studi Kasus

Kita ingin mengetahui:

* Apakah jam belajar memengaruhi nilai?
* Kategori jam belajar mana yang paling efektif?

---

### ğŸ” Challenge Project

1. Generate 50 data siswa
2. Buat DataFrame
3. Hitung:

   * Rata-rata nilai
   * Nilai tertinggi & terendah
4. Filter siswa dengan nilai â‰¥ 80
5. Visualisasikan:

   * Scatter jam belajar vs nilai
   * Bar chart rata-rata nilai per kategori

---

### âœ… Jawaban Challenge (Full Code)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(1)

# 1. Generate data
jam_belajar = np.random.randint(1, 11, size=50)
nilai = jam_belajar * 7 + np.random.randint(-8, 8, size=50)

# 2. DataFrame
df = pd.DataFrame({
    "Jam_Belajar": jam_belajar,
    "Nilai": nilai
})

# 3. Analisis
print("Rata-rata Nilai:", df["Nilai"].mean())
print("Nilai Tertinggi:", df["Nilai"].max())
print("Nilai Terendah:", df["Nilai"].min())

# 4. Filtering
df_lulus = df[df["Nilai"] >= 80]

# 5. Kategori
df["Kategori"] = pd.cut(
    df["Jam_Belajar"],
    bins=[0, 3, 6, 10],
    labels=["Sedikit", "Sedang", "Banyak"]
)

rata_kategori = df.groupby("Kategori")["Nilai"].mean()

# Visualisasi
plt.figure()
plt.scatter(df["Jam_Belajar"], df["Nilai"])
plt.xlabel("Jam Belajar")
plt.ylabel("Nilai")
plt.title("Jam Belajar vs Nilai")
plt.show()

plt.figure()
rata_kategori.plot(kind="bar")
plt.xlabel("Kategori Jam Belajar")
plt.ylabel("Rata-rata Nilai")
plt.title("Rata-rata Nilai per Kategori")
plt.show()
```
